```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import slic
from skimage.color import label2rgb
from skimage import graph
from skimage.segmentation import mark_boundaries
from collections import deque
import maxflow
import time
```

```python
scribble_mask = []
drawing = False
current_label = 1  # 1 for FG (red), 2 for BG (blue)
```

```python
def scribble_seeds(event, x, y, flags, param):
    global drawing, current_label, scale_x, scale_y

    img_x = int(x * scale_x)
    img_y = int(y * scale_y)

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        current_label = 1  # Foreground
    elif event == cv2.EVENT_RBUTTONDOWN:
        drawing = True
        current_label = 2  # Background
    elif event in [cv2.EVENT_LBUTTONUP, cv2.EVENT_RBUTTONUP]:
        drawing = False
    elif event == cv2.EVENT_MOUSEMOVE and drawing:
        if 0 <= img_x < scribble_mask.shape[1] and 0 <= img_y < scribble_mask.shape[0]:
            cv2.circle(scribble_mask, (img_x, img_y), 5, current_label, -1)


def get_user_seeds(image, segments, max_display_size=800):
    '''
    Get initial seeds/inputted foreground and background information from user.
    '''
    global scribble_mask, scale_x, scale_y

    segments = segments.astype(np.int32)
    scribble_mask = np.zeros(image.shape[:2], dtype=np.uint8)

    # Calculate scale for display
    h, w = image.shape[:2]
    scale = min(max_display_size / h, max_display_size / w, 1.0)
    display_size = (int(w * scale), int(h * scale))
    scale_x = w / display_size[0]
    scale_y = h / display_size[1]

    cv2.namedWindow("Scribble on Image", cv2.WINDOW_NORMAL)
    cv2.resizeWindow("Scribble on Image", *display_size)
    cv2.setMouseCallback("Scribble on Image", scribble_seeds)

    while True:
        overlay = image.copy()
        overlay[scribble_mask == 1] = [0, 0, 255]   # Red for foreground
        overlay[scribble_mask == 2] = [255, 0, 0]   # Blue for background
        overlay_resized = cv2.resize(overlay, display_size, interpolation=cv2.INTER_NEAREST)

        cv2.imshow("Scribble on Image", overlay_resized)
        key = cv2.waitKey(1)
        if key == 13: # Enter Key
            break

    cv2.destroyAllWindows()

    fg_seeds = set(np.unique(segments[scribble_mask == 1]))
    bg_seeds = set(np.unique(segments[scribble_mask == 2]))
    fg_seeds.discard(0)
    bg_seeds.discard(0)

    return fg_seeds, bg_seeds
```

```python
def mean_shift(image):
    '''
    Initial low level segmentation to speed up graph cut process. Uses SLIC (superpixel clustering)
    from skimage for faster segmentation than the mean shift in OpenCV.
    '''
    return slic(image, n_segments=500, compactness=10)


def build_graph(image: np.ndarray, segments: np.ndarray) -> graph.RAG:
    """
    Build a Region Adjacency Graph (RAG) based on mean color similarity.

    Args:
        image (np.ndarray): Input image as a (H, W, 3) RGB array.
        segments (np.ndarray): Segmentation map as a (H, W) array of integer labels for superpixel regions.

    Returns:
        skimage.future.graph.RAG: The constructed Region Adjacency Graph with nodes as region labels.
    """
    return graph.rag_mean_color(image, segments, mode='similarity')


def build_region_models(image: np.ndarray, segments: np.ndarray, fg_seeds: set[int], bg_seeds: set[int]):
    """
    Estimate foreground/background color models (mean RGB per region).

    Returns:
        mean_colors (dict[int, np.ndarray]): Mean RGB color for each region.
        fg_model (np.ndarray): Average RGB color of foreground seed regions.
        bg_model (np.ndarray): Average RGB color of background seed regions.
    """
    mean_colors = {}
    for label in np.unique(segments):
        mask = segments == label
        mean_colors[label] = np.mean(image[mask], axis=0)

    fg_model = np.mean([mean_colors[label] for label in fg_seeds], axis=0)
    bg_model = np.mean([mean_colors[label] for label in bg_seeds], axis=0)

    return mean_colors, fg_model, bg_model


def compute_data_term(region_color: np.ndarray, fg_model: np.ndarray, bg_model: np.ndarray):
    """
    Compute the unary data term using squared Euclidean distance.
    """
    fg_cost = np.linalg.norm(region_color - fg_model) ** 2
    bg_cost = np.linalg.norm(region_color - bg_model) ** 2
    return fg_cost, bg_cost


def get_adjacent_regions(rag: graph.RAG, regions: set[int]) -> set[int]:
    """
    Get all regions adjacent to the current set of regions in the Region Adjacency Graph (RAG).
    """
    neighbors = set()
    for r in regions:
        neighbors |= set(rag.neighbors(r))
    return neighbors - regions


def compute_smoothness_weight(color1: np.ndarray, color2: np.ndarray, delta: float = 25.0, distance: float = 1.0) -> float:
    """
    Compute Gaussian-based smoothness term for edge weight, encouraging similar regions to stay connected.

    Args:
        color1, color2 (np.ndarray): Mean RGB colors of the two regions.
        delta (float): Controls sensitivity to color difference.
        distance (float): Spatial distance (default=1 for adjacent regions).

    Returns:
        float: Edge weight between regions.
    """
    color_diff = np.linalg.norm(color1 - color2)
    weight = np.exp(-(color_diff ** 2) / (2 * delta ** 2)) / distance
    return weight


def iterated_graph_cut(image, segments, rag, fg_seeds, bg_seeds, max_iters=10):
    """
    Iterated Graph Cuts based on Peng et al. Algorithm 1.
    
    Parameters:
        image (np.ndarray): RGB image.
        segments (np.ndarray): Superpixel label image.
        rag (RAG): Region Adjacency Graph (from skimage.future.graph).
        fg_seeds (set[int]): Initial foreground region labels.
        bg_seeds (set[int]): Initial background region labels.
        max_iters (int): Maximum number of refinement iterations.

    Returns:
        fg_seeds (set[int]): Final foreground region labels.
        final_mask (np.ndarray): Binary segmentation mask (1=foreground, 0=background).
    """
    # Step 1: Add adjacent regions of foreground regions into G
    fg_seeds = set(fg_seeds)
    bg_seeds = set(bg_seeds)
    R = fg_seeds | bg_seeds
    G = R | get_adjacent_regions(rag, fg_seeds)

    # Step 2: Construct foreground and background data models from seed regions R
    region_colors, _, _ = build_region_models(image, segments, fg_seeds, bg_seeds)
    num_regions = len(region_colors)

    for _ in range(max_iters):
        region_colors, fg_model, bg_model = build_region_models(image, segments, fg_seeds, bg_seeds)

        # Step 3: build graph
        g = maxflow.Graph[float](num_regions, num_regions * 4)
        node_ids = {}
        for label in G:
            node_ids[label] = g.add_nodes(1)

        for label in G:
            fg_cost, bg_cost = compute_data_term(region_colors[label], fg_model, bg_model)
            g.add_tedge(node_ids[label], float(fg_cost), float(bg_cost))

        for label in G:
            for neighbor in rag.neighbors(label):
                if neighbor not in G:
                    continue
                w = compute_smoothness_weight(region_colors[label], region_colors[neighbor], delta=25.0)
                g.add_edge(node_ids[label], node_ids[neighbor], float(w), float(w))

        # Step 4: graph cut
        g.maxflow()
        new_fg = set()
        new_bg = set()
        for label in G:
            if int(g.get_segment(node_ids[label])) == 0:
                new_fg.add(label)
            else:
                new_bg.add(label)

        # Step 4 continued: update seed sets
        prev_len = len(fg_seeds)
        fg_seeds |= new_fg
        bg_seeds |= new_bg
        R = fg_seeds | bg_seeds

        # Step 5: expand working graph
        new_G = R | get_adjacent_regions(rag, fg_seeds)
        if new_G == G:  # convergence
            break
        G = new_G

    # Step 7: assign labels
    final_mask = np.zeros_like(segments, dtype=np.uint8)
    for label in fg_seeds:
        final_mask[segments == label] = 1
    return fg_seeds, final_mask
```

```python
def segments_to_mask(segments, fg_labels):
    mask = np.zeros_like(segments, dtype=np.uint8)
    for label in fg_labels:
        mask[segments == label] = 255
    return mask
```
