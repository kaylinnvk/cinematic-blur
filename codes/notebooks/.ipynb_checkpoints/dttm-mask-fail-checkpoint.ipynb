```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import scipy.fft as fft
from numpy.fft import rfft2, irfft2
import time
```

```python
# GLOBAL SETUPS
TEMPLATE_BANK_SIZE = 5
NCC_THRESHOLD = 0.5
template_bank = []
```

##
## Utility Functions

```python
def read_video_first_frame(video_path):
    """Open video capture and read the first frame."""
    cap = cv2.VideoCapture(video_path)
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Cannot read video.")
        cap.release()
        return None, None, False
    return cap, frame, True

def load_template(initial_template_path, frame_gray):
    """
    Load grayscale template image and crop the template area from the given grayscale frame.
    Returns cropped template and its coordinates in the frame.
    """
    template = cv2.imread(initial_template_path, cv2.IMREAD_GRAYSCALE)
    if template is None:
        print("‚ùå Template image not found or failed to load.")
        return None, None

    coords = cv2.findNonZero(template)
    if coords is None:
        print("‚ùå No non-zero pixels found in template.")
        return None, None

    x, y, w, h = cv2.boundingRect(coords)
    cropped_template = frame_gray[y:y+h, x:x+w]
    return cropped_template, (x, y, w, h)

def setup_video_writer(output_path, frame_shape, fps):
    """Initialize and return a VideoWriter object."""
    height, width = frame_shape[:2]
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
    print(f"üíæ Output video will be saved to: {output_path}")
    return out
```

##
## Masked FFT NCC

```python
def precompute_template_fft_rfft(template, mask, search_patch_shape):
    """
    Precompute and cache real FFTs needed for masked FFT NCC for a given template and mask.

    Parameters:
        template (ndarray): template image (H2, W2)
        mask (ndarray): binary mask same size as template
        search_patch_shape (tuple): (H1, W1) shape of the search patch (fixed image)

    Returns:
        dict with FFTs and padding info to reuse in matching
    """
    H1, W1 = search_patch_shape
    H2, W2 = template.shape
    H = H1 + H2 - 1
    W = W1 + W2 - 1

    def pad(img):
        return np.pad(img, ((0, H - img.shape[0]), (0, W - img.shape[1])), mode='constant')

    f2 = pad(template * mask)
    m2 = pad(mask)
    f2_sq = pad((template ** 2) * mask)

    F2 = rfft2(f2)
    M2 = rfft2(m2)
    F2_sq = rfft2(f2_sq)

    sum_f2 = np.sum(template * mask)

    return {
        'F2': F2,
        'M2': M2,
        'F2_sq': F2_sq,
        'H': H,
        'W': W,
        'template_shape': (H2, W2),
        'sum_f2': sum_f2
    }


def masked_fft_ncc_cached_rfft(fixed, cache_dict, min_overlap_ratio=0.1):
    """
    Masked FFT NCC using cached real FFTs for the template and mask.

    Parameters:
        fixed (ndarray): search patch (fixed image)
        cache_dict (dict): cached FFTs and parameters from precompute_template_fft_rfft
        min_overlap_ratio (float): minimum overlap ratio threshold

    Returns:
        ncc_valid (ndarray): NCC response map
        (dy, dx): location of best match in the search patch
    """
    fixed = fixed.astype(np.float32)

    H1, W1 = fixed.shape
    H2, W2 = cache_dict['template_shape']
    H, W = cache_dict['H'], cache_dict['W']

    def pad(img):
        return np.pad(img, ((0, H - img.shape[0]), (0, W - img.shape[1])), mode='constant')

    f1 = pad(fixed)
    f1_sq = pad(fixed ** 2)

    F1 = rfft2(f1)
    F1_sq = rfft2(f1_sq)

    F2 = cache_dict['F2']
    M2 = cache_dict['M2']
    F2_sq = cache_dict['F2_sq']

    # Cross-correlations (inverse FFT of product of F1 and conj(F2))
    numerator = irfft2(F1 * np.conj(F2), s=(H, W))
    sum_f1 = irfft2(F1 * np.conj(M2), s=(H, W))

    overlap = irfft2(M2, s=(H, W))
    sum_f2 = cache_dict['sum_f2']

    sum_f1_sq = irfft2(F1_sq * np.conj(M2), s=(H, W))
    sum_f2_sq = irfft2(F2_sq, s=(H, W))

    eps = 1e-10
    numerator -= (sum_f1 * sum_f2) / (overlap + eps)
    denom = np.sqrt(np.maximum(sum_f1_sq * sum_f2_sq, 0))
    ncc = numerator / (denom + eps)

    overlap_thresh = min_overlap_ratio * np.max(overlap)
    ncc[overlap < overlap_thresh] = 0

    y0 = H2 - 1
    x0 = W2 - 1
    y1 = y0 + (H1 - H2 + 1)
    x1 = x0 + (W1 - W2 + 1)
    ncc_valid = ncc[y0:y1, x0:x1]

    dy, dx = np.unravel_index(np.argmax(ncc_valid), ncc_valid.shape)

    return ncc_valid, (dy, dx)
```

##
## Template Matching + Bank

```python
def get_patch(frame_gray, template_coords, padding=10):
    """Extract a patch from the grayscale frame around the template coordinates with padding."""
    x, y, w, h = template_coords
    x_start = max(0, x - padding)
    y_start = max(0, y - padding)
    x_end = min(frame_gray.shape[1], x + w + padding)
    y_end = min(frame_gray.shape[0], y + h + padding)
    patch = frame_gray[y_start:y_end, x_start:x_end]
    patch_loc = (x_start, y_start)
    return patch, patch_loc

def display_patch(patch, frame_idx, padding):
    """Display a grayscale image patch for debugging or visualization."""
    plt.figure(figsize=(4, 4))
    plt.imshow(patch, cmap='gray')
    plt.title(f"Patch (Frame {frame_idx}, Padding {padding})")
    plt.axis('off')
    plt.show()

def find_best_match(search_patch):
    global template_bank
    best_score = -1
    best_box = None
    boxes = []
    scores = []

    for entry in template_bank:
        cache = entry['cache']
        ncc_map, max_loc = masked_fft_ncc_cached_rfft(search_patch, cache)
        max_val = np.max(ncc_map)

        if max_val == -1:
            continue

        if max_val > best_score:
            best_score = max_val
            best_box = (*max_loc, entry['template'].shape[1], entry['template'].shape[0])

        boxes.append((*max_loc, entry['template'].shape[1], entry['template'].shape[0]))
        scores.append(max_val)

    if best_box is None:
        best_box = (0, 0, 0, 0)

    return boxes, scores, best_score

def update_template_from_box(frame_gray, box):
    x, y, w, h = box
    template_crop = frame_gray[y:y+h, x:x+w]
    _, mask = cv2.threshold(template_crop, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return template_crop, mask

def add_to_template_bank(patch, patch_mask, search_patch_shape, max_size=TEMPLATE_BANK_SIZE):
    """
    Add new template and mask to global template_bank with cached FFTs.

    search_patch_shape = shape of the search patch you expect to use for matching, e.g. (H, W)
    """
    global template_bank
    cache = precompute_template_fft_rfft(patch, patch_mask, search_patch_shape)
    template_bank.append({
        'template': patch,
        'mask': patch_mask,
        'cache': cache
    })
    if len(template_bank) > max_size:
        template_bank.pop(0)

def apply_nms(boxes, scores, threshold, nms_thresh):
    """
    Apply Non-Maximum Suppression to filter overlapping bounding boxes.
    Returns list of indices of boxes to keep.
    """
    if not boxes:
        return []

    indices = cv2.dnn.NMSBoxes(boxes, scores, threshold, nms_thresh)
    if len(indices) == 0:
        return []

    return [i[0] if isinstance(i, (list, np.ndarray)) else i for i in indices]

def visualize_template_bank(template_bank, current_index=None, max_per_row=5, figsize=(15, 5), cmap='gray'):
    """Visualize all templates in the bank, highlighting the current one."""
    n = len(template_bank)
    if n == 0:
        print("‚ö†Ô∏è Template bank is empty.")
        return

    cols = min(n, max_per_row)
    rows = (n + cols - 1) // cols

    plt.figure(figsize=figsize)
    for i, entry in enumerate(template_bank):
        template = entry['template']
        mask = entry['mask']

        ax = plt.subplot(rows, cols, i + 1)

        # Highlight masked region: show masked template only
        masked_template = cv2.bitwise_and(template, template, mask=mask)

        # Optional: overlay red mask (for visualization clarity)
        overlay = np.stack([masked_template] * 3, axis=-1)  # convert to RGB
        red_mask = (mask == 0)
        overlay[red_mask] = [255, 0, 0]  # red background

        ax.imshow(overlay)
        title = f"Template {i}"
        if i == current_index:
            title += " (Current)"
            ax.set_title(title, color='red')
            for spine in ax.spines.values():
                spine.set_edgecolor('red')
                spine.set_linewidth(2)
        else:
            ax.set_title(title)
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

##
## Main Tracking Functions

```python
def draw_tracking_box(frame, box):
    """Draw a rectangle tracking box on the frame."""
    x, y, w, h = box
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)

def display_frame(frame, frame_idx):
    """Display the current video frame in RGB with a title."""
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    clear_output(wait=True)
    plt.figure(figsize=(10, 5))
    plt.imshow(frame_rgb)
    plt.title(f"Frame {frame_idx}")
    plt.axis('off')
    plt.show()

def template_difference(template1, template2, threshold=0.2):
    """
    Compute normalized mean squared error difference between two templates.
    Returns True if difference is greater than threshold.
    """
    if template1.shape != template2.shape:
        return True  # Different shapes are considered different

    diff = np.mean((template1.astype(np.float32) - template2.astype(np.float32)) ** 2)
    norm = np.mean(template1.astype(np.float32) ** 2)
    nmse = diff / (norm + 1e-8)
    return nmse > threshold

def dynamic_template_matching(video_path, initial_template_path,
                              output_path="../../outputs/dttm_output.mp4",
                              threshold=0.7, nms_thresh=0.3):
    """
    Perform dynamic template matching on video frames, updating template bank and writing output video.
    """
    global template_bank
    
    padding = 10
    max_padding = 50
    min_padding = 5
    padding_step = 5
    
    cap, frame, ret = read_video_first_frame(video_path)
    if not ret:
        return

    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    template, template_coords = load_template(initial_template_path, frame_gray)
    template_img = cv2.imread(initial_template_path, cv2.IMREAD_GRAYSCALE)
    if template is None:
        cap.release()
        return
    
    # Create initial mask that maps all nonzero values to 1 
    initial_mask = (template > 0).astype(np.uint8)

    # Compute search_patch_shape based on padding you plan to use:
    search_patch_shape = get_patch(frame_gray, template_coords, padding=10)[0].shape

    add_to_template_bank(template, initial_mask, search_patch_shape)

    fps = cap.get(cv2.CAP_PROP_FPS) or 30
    out = setup_video_writer(output_path, frame.shape, fps)

    frame_idx = 0

    while True:
        if frame_idx != 0:
            ret, frame = cap.read()
            if not ret:
                print("üìΩÔ∏è Video ended.")
                break
            frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        search_patch, patch_loc = get_patch(frame_gray, template_coords, padding)

        # Uncomment below line to visualize patches for debugging
        # display_patch(search_patch, frame_idx, padding)

        boxes, scores, max_val = find_best_match(search_patch)
        print(f"[Frame {frame_idx}] Max match score: {max_val:.3f}")

        indices = apply_nms(boxes, scores, threshold, nms_thresh)
        if indices:
            padding = max(min_padding, padding - padding_step)
            best_idx = indices[0]
            best_box = boxes[best_idx]

            px, py = patch_loc
            global_box = (best_box[0] + px, best_box[1] + py, best_box[2], best_box[3])
            print(f"  ‚Ü™ Object tracked at {global_box}")
            draw_tracking_box(frame, global_box)

            new_template, new_mask = update_template_from_box(frame_gray, global_box)
            if new_template is not None and template_difference(new_template, template_img, threshold=0.2):
                template_img = new_template
                template_coords = global_box
                add_to_template_bank(template_img, new_mask)
                print("  ‚Ü™ Template updated and added to bank.\n")
            else:
                print("  ‚Ü™ Template not different enough, skipped.\n")
        else:
            padding = min(max_padding, padding + padding_step)
            print("  ‚ö†Ô∏è No match above threshold or NMS removed all boxes.\n")

        # Uncomment below lines to visualize template bank
        # current_template_index = get_template_index(template, template_bank)
        # visualize_template_bank(template_bank, current_index=current_template_index)

        out.write(frame)
        display_frame(frame, frame_idx)
        frame_idx += 1

    cap.release()
    out.release()
    print("üé• Video saved successfully.")
```

```python
def visualize_masked_template(template, mask, title="Masked Template"):
    """Display the template with its mask applied and background in red."""
    masked = cv2.bitwise_and(template, template, mask=mask)

    # Convert to 3-channel for colored visualization
    overlay = np.stack([masked] * 3, axis=-1)
    red_bg = (mask == 0)
    overlay[red_bg] = [255, 0, 0]  # Red for background

    plt.figure(figsize=(4, 4))
    plt.imshow(overlay)
    plt.title(title)
    plt.axis('off')
    plt.show()
```
