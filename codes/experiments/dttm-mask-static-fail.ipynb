```python
def load_template(initial_template_path, frame_gray):
    template = cv2.imread(initial_template_path, cv2.IMREAD_GRAYSCALE)
    if template is None:
        print("Template image not found or failed to load.")
        return None

    coords = cv2.findNonZero(template)
    if coords is None:
        print("No non-zero pixels in template.")
        return None

    x, y, w, h = cv2.boundingRect(coords)
    cropped_template = frame_gray[y:y+h, x:x+w]
    return cropped_template, (x, y, w, h)

def setup_video_writer(output_path, frame_shape, fps):
    frame_h, frame_w = frame_shape[:2]
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_w, frame_h))
    print(f"üíæ Output video will be saved to: {output_path}")
    return out

```

```python
def masked_fft_ncc(search, template, mask_template):
    """
    Compute masked normalized cross-correlation using FFT based on the formula you gave.

    Args:
        search: Grayscale search image patch, float32 np.array (H, W)
        template: Grayscale template image, float32 np.array (h, w)
        mask_template: Binary mask (0/1) for the template, same shape as template

    Returns:
        ncc_map: correlation map (H - h + 1, W - w + 1)
        max_loc: (x, y) location of best match in search coords
    """
    # Convert to float32 if needed
    search = search.astype(np.float32)
    template = template.astype(np.float32)
    mask_template = mask_template.astype(np.float32)

    H, W = search.shape
    h, w = template.shape

    # Pad sizes for FFT (optimal DFT sizes)
    fft_h = cv2.getOptimalDFTSize(H + h - 1)
    fft_w = cv2.getOptimalDFTSize(W + w - 1)

    # Zero-pad inputs
    f1 = np.zeros((fft_h, fft_w), dtype=np.float32)
    f1[:H, :W] = search * 1.0  # f1 = search * mask1 (mask1 = all ones here)

    m1 = np.ones_like(f1, dtype=np.float32)  # mask of search patch, here all ones

    f2 = np.zeros((fft_h, fft_w), dtype=np.float32)
    f2[:h, :w] = template * mask_template

    m2 = np.zeros((fft_h, fft_w), dtype=np.float32)
    m2[:h, :w] = mask_template

    # Compute FFTs
    F1 = cv2.dft(f1, flags=cv2.DFT_COMPLEX_OUTPUT)
    F2 = cv2.dft(f2, flags=cv2.DFT_COMPLEX_OUTPUT)
    M1 = cv2.dft(m1, flags=cv2.DFT_COMPLEX_OUTPUT)
    M2 = cv2.dft(m2, flags=cv2.DFT_COMPLEX_OUTPUT)

    def conj(b):
        """Compute complex conjugate of DFT b."""
        b_conj = b.copy()
        b_conj[:, :, 1] *= -1
        return b_conj

    # Helper to do inverse DFT real output
    def idft_real(c):
        return cv2.idft(c, flags=cv2.DFT_SCALE | cv2.DFT_REAL_OUTPUT)

    # Cross-correlation numerator (equation (12) masked)
    numerator = idft_real(cv2.mulSpectrums(F1, conj(F2), flags=0)) \
                - (idft_real(cv2.mulSpectrums(F1, conj(M2), flags=0)) *
                   idft_real(cv2.mulSpectrums(M1, conj(F2), flags=0))) / (idft_real(cv2.mulSpectrums(M1, conj(M2), flags=0)) + 1e-10)

    # Denominator parts (equations (13) and (14))
    den1 = idft_real(cv2.mulSpectrums(cv2.dft(f1 * f1, flags=cv2.DFT_COMPLEX_OUTPUT), conj(M2), flags=0)) \
           - (idft_real(cv2.mulSpectrums(F1, conj(M2), flags=0)) ** 2) / (idft_real(cv2.mulSpectrums(M1, conj(M2), flags=0)) + 1e-10)

    f2_zero_mean = f2 - (idft_real(cv2.mulSpectrums(M1, conj(F2), flags=0)) / (idft_real(cv2.mulSpectrums(M1, conj(M2), flags=0)) + 1e-10))
    den2 = idft_real(cv2.mulSpectrums(M1, cv2.dft(f2_zero_mean * f2_zero_mean, flags=cv2.DFT_COMPLEX_OUTPUT), flags=0)) \
           - (idft_real(cv2.mulSpectrums(M1, conj(F2), flags=0)) ** 2) / (idft_real(cv2.mulSpectrums(M1, conj(M2), flags=0)) + 1e-10)

    # Final denominator
    denominator = np.sqrt(np.maximum(den1 * den2, 1e-10))

    # Valid region (size: H - h + 1, W - w + 1)
    valid_h = H - h + 1
    valid_w = W - w + 1

    ncc_map = numerator[h-1:H, w-1:W] / denominator[h-1:H, w-1:W]

    # Clamp values to [-1, 1] to avoid numerical issues
    ncc_map = np.clip(ncc_map, -1, 1)

    # Find best match
    _, max_val, _, max_loc = cv2.minMaxLoc(ncc_map.astype(np.float32))

    # After computing numerator and denominator
    numerator = ncc_num
    denominator = np.sqrt(np.maximum(den1 * den2, 1e-10))
    full_ncc = numerator / denominator

    # ‚úÖ Visualize full NCC result (before cropping valid region)
    visualize_ncc_map(full_ncc, title="Full NCC Map (Uncropped)")

    return ncc_map, max_loc
```

```python
def get_patch(frame_gray, template_coords, padding=10):
    x, y, w, h = template_coords
    x_start = max(0, x - padding)
    y_start = max(0, y - padding)
    x_end = min(frame_gray.shape[1], x + w + padding)
    y_end = min(frame_gray.shape[0], y + h + padding)

    patch = frame_gray[y_start:y_end, x_start:x_end]
    patch_loc = (x_start, y_start)  # top-left of patch

    return patch, patch_loc

def display_patch(patch, frame_idx, padding):
    plt.figure(figsize=(4, 4))
    plt.imshow(patch, cmap='gray')
    plt.title(f"Patch (Frame {frame_idx}, Padding {padding})")
    plt.axis('off')
    plt.show()

def find_best_match(search_patch):
    global template_bank
    
    best_score = -1
    best_box = None
    boxes = []
    scores = []

    for template in template_bank:
        if search_patch.shape[0] < template.shape[0] or search_patch.shape[1] < template.shape[1]:
            continue  # Skip if template is larger than the search patch

        result = cv2.matchTemplate(search_patch, template, cv2.TM_CCOEFF_NORMED)
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

        if max_val > best_score:
            best_score = max_val
            best_box = (*max_loc, template.shape[1], template.shape[0])  # (x, y, w, h)

        boxes.append((*max_loc, template.shape[1], template.shape[0]))
        scores.append(max_val)

    if best_box is None:
        best_box = (0, 0, 0, 0)

    return boxes, scores, best_score


def update_template_from_box(frame_gray, box):
    x, y, w, h = box
    new_template = frame_gray[y:y+h, x:x+w]
    coords = cv2.findNonZero(new_template)
    if coords is not None:
        x_, y_, w_, h_ = cv2.boundingRect(coords)
        return new_template[y_:y_+h_, x_:x_+w_]
    return new_template

def add_to_template_bank(patch):
    global template_bank
    
    template_bank.append(patch)
    if len(template_bank) > TEMPLATE_BANK_SIZE:
        template_bank.pop(0)

def apply_nms(boxes, scores, threshold, nms_thresh):
    if not boxes:
        return []

    indices = cv2.dnn.NMSBoxes(boxes, scores, threshold, nms_thresh)
    if len(indices) == 0:
        return []

    # Flatten indices to list of integers
    return [i[0] if isinstance(i, (list, np.ndarray)) else i for i in indices]

#================================================#
#======= FOR TEMPLATE BANK VISUALIZATION ========#
#================================================#

def get_template_index(template, template_bank):
    for i, t in enumerate(template_bank):
        if np.array_equal(t, template):
            return i
    return None  # Not found

def visualize_template_bank(template_bank, current_index=None, max_per_row=5, figsize=(15, 5), cmap='gray'):
    n = len(template_bank)
    if n == 0:
        print("‚ö†Ô∏è Template bank is empty.")
        return

    cols = min(n, max_per_row)
    rows = (n + cols - 1) // cols

    plt.figure(figsize=figsize)
    for i, template in enumerate(template_bank):
        ax = plt.subplot(rows, cols, i + 1)
        plt.imshow(template, cmap=cmap)
        title = f"Template {i}"
        if i == current_index:
            # Highlight current template with red title and border
            title += " (Current)"
            ax.set_title(title, color='red')
            for spine in ax.spines.values():
                spine.set_edgecolor('red')
                spine.set_linewidth(2)
        else:
            ax.set_title(title)
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def draw_tracking_box(frame, box):
    x, y, w, h = box
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)

def display_frame(frame, frame_idx):
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    clear_output(wait=True)
    plt.figure(figsize=(10, 5))
    plt.imshow(frame_rgb)
    plt.title(f"Frame {frame_idx}")
    plt.axis('off')
    plt.show()

def template_difference(template1, template2, threshold=0.2):
    if template1.shape != template2.shape:
        return True  # treat different shape as significantly different
    diff = np.mean((template1.astype(np.float32) - template2.astype(np.float32)) ** 2)
    norm = np.mean(template1.astype(np.float32) ** 2)
    nmse = diff / (norm + 1e-8)
    return nmse > threshold
```

```python
def visualize_ncc_map(ncc_map, title="Full NCC Map"):
    plt.figure(figsize=(6, 6))
    plt.imshow(ncc_map, cmap='jet')
    plt.colorbar()
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def static_template_match(image_path, initial_template_path, padding=10, threshold=0.7):
    template_bank = []  # Reset for static test

    # Load input image
    frame = cv2.imread(image_path)
    if frame is None:
        print("‚ùå Failed to load image.")
        return

    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Load and extract template using nonzero-mask bounding box
    template, template_coords = load_template(initial_template_path, frame_gray)
    if template is None:
        print("‚ùå Template could not be loaded.")
        return

    add_to_template_bank(template)

    # Get the search region based on template coords
    search_patch, patch_loc = get_patch(frame_gray, template_coords, padding)

    # Find best match using your NCC function
    boxes, scores, max_val = find_best_match(search_patch)
    print(f"üîç Max NCC match score: {max_val:.3f}")

    # Apply NMS if desired
    indices = apply_nms(boxes, scores, threshold, nms_thresh=0.3)
    if indices:
        best_idx = indices[0]
        best_box = boxes[best_idx]
        px, py = patch_loc
        global_box = (best_box[0] + px, best_box[1] + py, best_box[2], best_box[3])

        # Draw result
        draw_tracking_box(frame, global_box)
        display_frame(frame, frame_idx=0)
        print(f"‚úÖ Match found at {global_box} with score {scores[best_idx]:.3f}")
    else:
        print("‚ö†Ô∏è No match above threshold or suppressed by NMS.")
```

```python
def compare_ncc_methods(search_patch, template, mask_template):
    # Ensure float32 and scale to [0,1]
    search = search_patch.astype(np.float32) / 255.0
    templ = template.astype(np.float32) / 255.0
    mask = mask_template.astype(np.float32)

    # --- Regular cv2.matchTemplate NCC ---
    regular_result = cv2.matchTemplate(search, templ, cv2.TM_CCOEFF_NORMED)
    max_val_regular = regular_result.max()

    # --- Masked FFT NCC (your function) ---
    ncc_map_masked, max_loc_masked = masked_fft_ncc(search, templ, mask)
    max_val_masked = ncc_map_masked.max()

    # --- Visualization ---
    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.title("Search Patch")
    plt.imshow(search, cmap='gray')
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.title("Template (Masked)")
    plt.imshow(templ, cmap='gray')
    plt.imshow(mask, cmap='jet', alpha=0.3)  # Overlay mask
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.title(f"NCC Maps\nRegular max={max_val_regular:.3f}, Masked max={max_val_masked:.3f}")
    plt.plot(regular_result.ravel(), label='Regular NCC', alpha=0.7)
    plt.plot(ncc_map_masked.ravel(), label='Masked FFT NCC', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"Max regular NCC score: {max_val_regular:.3f}")
    print(f"Max masked FFT NCC score: {max_val_masked:.3f}")

    return max_val_regular, max_val_masked

# Example usage (replace with your actual data):
# max_reg, max_masked = compare_ncc_methods(search_patch, template, mask_template)
```
